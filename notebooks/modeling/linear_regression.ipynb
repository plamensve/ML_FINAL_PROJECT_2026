## Linear Regression for Football Player Salary Prediction and Market Valuation

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np
from sympy.benchmarks.bench_discrete_log import data_set_2
```

```python
attributes = pd.read_csv('../../data/processed/personal_performance_data.csv')
target = pd.read_csv('../../data/processed/wage_euro_target_feature.csv')
```

```python
attributes.isna().sum()
```

```python
target.isna().sum()
```

```python
target = target.fillna(target.mean())
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    attributes, target, random_state=42, test_size=0.1
)
```

```python
# Create the Linear Regression model.
model = LinearRegression()
```

```python
# Train the model using the training data.
model.fit(X_train, y_train)
```

```python
# Predict the target values for the test data.
y_pred = model.predict(X_test)
```

```python
# Calculate regression evaluation metrics.
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print(f"MAE: {mae:.2f}")
print(f"R2 Score: {r2:.4f}")
```

## Linear Regression Model Evaluation

The Linear Regression model achieved a Mean Absolute Error (MAE) of **9,087.33 euros** and an R² score of **0.2976**.

This means that, on average, the model makes an error of around **9,087 euros** when predicting player wages. The R² score shows that the model explains approximately **29.76%** of the variation in the target variable.

Overall, the model has limited predictive performance. It captures some relationship between the selected features and the target, but a large part of the wage variation remains unexplained.

To improve the model, more relevant information should be added. Player wages are not determined only by personal performance statistics. They can also depend on factors such as player reputation, age, club level, league quality, contract status, market value, position, nationality, and overall rating. Adding more informative features could help the model make more accurate predictions.

```python
data_set_2  = pd.read_csv('../../data/processed/clustering_personal_performance_data.csv')
main_dataset = pd.read_csv('../../data/raw/fifa_players.csv')
```

```python
main_dataset
```

To improve the model performance, we will add several additional features that may have a stronger relationship with player wage prediction.

The new features will include:

- `age` - player age
- `overall_rating` - current overall player rating
- `potential` - estimated future potential
- `value_euro` - estimated market value in euros
- `international_reputation` - international reputation
- `clusters` - we will use dataset with the added clusters

These features are expected to provide more useful information to the model, because player wages are usually influenced not only by personal performance statistics, but also by overall quality, market value, age, and future potential.

```python
additional_features = main_dataset[['age', 'overall_rating', 'potential', 'value_euro', 'international_reputation(1-5)']]
```

```python
improved_dataset = pd.concat([data_set_2, additional_features], axis=1)
improved_dataset
```

```python
improved_dataset.isna().sum()
```

```python
pp_clusters_ohe = pd.get_dummies(
    improved_dataset['pp_cluster'],
    prefix='pp_cluster',
    dtype=int
)
improved_dataset = pd.concat([improved_dataset, pp_clusters_ohe], axis=1)
improved_dataset = improved_dataset.drop(columns=['pp_cluster'])
```

```python
improved_dataset
```

```python
improved_dataset.isna().sum()
```

```python
improved_dataset['value_euro'] = improved_dataset['value_euro'].fillna(improved_dataset['value_euro'] .mean())
```

```python
X_train, X_test, y_train, y_test = train_test_split(improved_dataset, target, random_state=42, train_size=0.25)
```

```python
model_2 = LinearRegression()
```

```python
model_2.fit(X_train, y_train)
```

```python
# Predict the target values for the test data.
y_pred = model_2.predict(X_test)

# Calculate regression evaluation metrics.
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print(f"MAE: {mae:.2f}")
print(f"R2 Score: {r2:.4f}")
```

## Improved Linear Regression Model Evaluation

After adding additional features to the dataset, the Linear Regression model performance improved significantly.

The updated model achieved a **Mean Absolute Error (MAE) of 4,869.35 euros** and an **R² score of 0.7618**.

Compared to the initial model, where the MAE was **9,087.33 euros** and the R² score was **0.2976**, the new model performs much better. The average prediction error decreased by approximately **4,218 euros**, and the model now explains around **76.18%** of the variation in player wages.

This improvement shows that the added features provide much stronger information for wage prediction. Features such as `age`, `overall_rating`, `potential`, `value_euro`, and `international_reputation` help the model better understand the factors that influence player wages.

The result also confirms that player wages are not determined only by personal performance statistics. They are also strongly influenced by the player's overall quality, market value, future potential, age, and international reputation.

Overall, the improved model has much better predictive performance and is significantly more suitable for wage prediction than the initial model.
